#**PyCaret Regression**

This script performs the following steps:

- It imports necessary libraries and fetches the California Housing dataset from sklearn.datasets.
- The dataset is then converted into a pandas DataFrame, which PyCaret uses to perform its operations.
- The setup function initializes the regression experiment with the dataset, specifying 'MedianHouseValue' as the target variable.
- compare_models() is used to train and evaluate different models, selecting the best model based on the default metric.
- Finally, plot_model is used twice to generate a learning curve and an actual vs. predicted values plot for the best-performing model.

```python
# Import necessary libraries
import numpy as np
import pandas as pd
from sklearn.datasets import fetch_california_housing
from pycaret.regression import *

# Fetch the California Housing dataset
california_housing = fetch_california_housing()
X = pd.DataFrame(california_housing.data, columns=california_housing.feature_names)
y = california_housing.target
data = pd.concat([X, pd.Series(y, name='MedianHouseValue')], axis=1)

# Initialize the setup for PyCaret
s = setup(data, target='MedianHouseValue', session_id=123)

# Compare models to find the best one based on default metrics
best_model = compare_models()

# Print the best model
print("Best Model:", best_model)

# Plot the learning curve of the best model
plot_model(best_model, plot='learning')

# Plot the Actual vs Predicted values
plot_model(best_model, plot='error')

```
