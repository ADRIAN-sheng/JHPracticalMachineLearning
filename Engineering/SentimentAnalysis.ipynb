#**SentimentAnalysis**

This code demonstrates sentiment analysis using:
- TextBlob, and
- A language model served by huggingface.com

The TextBlob sentiment property will give you a polarity score (ranging from -1 to 1, where 1 is positive sentiment) and a subjectivity score (ranging from 0 to 1, where 1 is more subjective). The Hugging Face transformers pipeline will provide a label (POSITIVE or NEGATIVE) and a confidence score.

The Hugging Face model requires internet access to download the pre-trained model the first time you run the sentiment analysis pipeline, and its performance may be significantly better and more nuanced than TextBlob's, especially for complex sentences or nuanced sentiments. ​

The huggingface model requires authentication using an API key at huggingface.com (accounts there are free)

##TextBlob

```python
import textblob
from textblob import TextBlob

# Sample text
text = "Holy shot dude! We talked about this yesterday! When are you going to get with the program? Do you have a brain?"
blob = TextBlob(text)

# Get the sentiment
textblob_sentiment = blob.sentiment
print(f"TextBlob Sentiment: {textblob_sentiment}")
```

##Huggingface Model (Language Model)

```python
#Authenticate huggingface
from huggingface_hub import notebook_login

notebook_login()
```

```python
from transformers import pipeline

# Load the sentiment analysis pipeline
transformer = pipeline('sentiment-analysis')

# Sample text
text = "I love this product. It works great!"

# Get the sentiment
transformer_sentiment = transformer(text)
print(f"Transformers Sentiment: {transformer_sentiment}")
```
