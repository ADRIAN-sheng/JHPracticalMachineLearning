**Random Forest Feature Importance**
Random Forest feature importance measures the significance of each feature in the decision making of the Random Forest models. It's based on how much each feature decreases the impurity of the split (e.g., Gini impurity for classification problems). Features that decrease the impurity the most are more important for making predictions.

This code does the following:
- Loads the Iris dataset.
- Initializes a RandomForestClassifier and fits it to the dataset.
- Extracts the feature importances from the trained model.
- Sorts the features by their importance in descending order.
- Creates a bar plot displaying the importance of each feature.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.ensemble import RandomForestClassifier

# Load the Iris dataset
data = load_iris()
X = data.data
y = data.target

# Initialize and train a RandomForestClassifier
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X, y)

# Calculate feature importances
feature_importances = model.feature_importances_

# Sort the feature importances in descending order and get the indices
sorted_idx = np.argsort(feature_importances)[::-1]

# Prepare labels in the order of importance
labels = np.array(data.feature_names)[sorted_idx]

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(range(X.shape[1]), feature_importances[sorted_idx], align='center')
plt.xticks(range(X.shape[1]), labels, rotation=45)
plt.xlabel('Feature Importance')
plt.ylabel('Relative Importance')
plt.title('Feature Importances in RandomForest Classifier')
plt.show()
```
